```python
#Looking at the data dictionary, it does not make sense to have the Year built be 0 so we will be replacing the value with the median value of the column
#Assistance from James Hampton , Seattle TA
                
#Create a method to convert the Garage Yr Blt to the appropriate decade
def Garage_Year(value, divisor):
    return (value - (value%divisor))

#applying the method to segregate the Garage Yr Blt by decade for easiler get_dummy columns while converting the NaN as NA
df_train['Garage Yr Blt'] = Garage_Year(df_train['Garage Yr Blt'],10)

#original data cleaning method : df_train['Garage Yr Blt'].fillna(0,inplace = True)
#new data cleaning method : fill NA using random variable that already exists in the column
df_train['Garage Yr Blt'].fillna(lambda x: random.choice(df_train[df_train[column] != np.nan]["Garage Yr Blt"]))

df_test['Garage Yr Blt'] = Garage_Year(df_test['Garage Yr Blt'],10)

#original data cleaning method : df_test['Garage Yr Blt'].fillna(0,inplace = True)
#new ata cleaning method : fill NA using random variable that already exists in the column
df_test['Garage Yr Blt'].fillna(lambda x: random.choice(df_test[df_test[column] != np.nan]["Garage Yr Blt"]))
```

```python
# using a for loop to drop the first rows of each of the get_dummy created columns for each set of feature that dummified its column
# credits to Boom

#for dummy in dummy_list:
#   for col in df_train.columns
#      if dummy in col
#     col_list.append(col)

dummy_list = nominal_list

for dummy in dummy_list:
    col_list = [col for col in df_train.columns if dummy in col]
    df_train.drop(columns=col_list[0],inplace = True)
    df_test.drop(columns= col_list[0],inplace = True)
    
```

```python
# Test whether the for loop for dummy list [0] removal works or not 
df_train.columns.unique()
```

```python
#Lasso does a sparse selection, while Ridge does not.
#When you have highly-correlated variables, Ridge regression shrinks the two coefficients towards one another. Lasso is somewhat indifferent and generally picks one over the other. Depending on the context, one does not know which variable gets picked. Elastic-net is a compromise between the two that attempts to shrink and do a sparse selection simultaneously.
#Ridge estimators are indifferent to multiplicative scaling of the data. That is, if both X and Y variables are multiplied by constants, the coefficients of the fit do not change, for a given λ parameter. However, for Lasso, the fit is not independent of the scaling. In fact, the λ parameter must be scaled up by the multiplier to get the same result. It is more complex for elastic net.
#Ridge penalizes the largest β's more than it penalizes the smaller ones (as they are squared in the penalty term). Lasso penalizes them more uniformly. This may or may not be important. In a forecasting problem with a powerful predictor, the predictor's effectiveness is shrunk by the Ridge as compared to the Lasso.
```

```python
#Calculating Baseline Score
base_mean = DummyRegressor(strategy = 'mean')


features = ['Overall Qual',
            'Gr Liv Area',
            'Total Bsmt SF',
            'Garage Area',  
            '1st Flr SF',   
            'Garage Cars',  
            'Bsmt Qual',    
            'Year Built',   
            'Garage Finish', 
            'Year Remod/Add']
X = df_train[features]
y = df_train['SalePrice']



#train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    random_state=42,
                                                    test_size=0.05)

base_mean.fit(X_train, y_train) 

y_hat_base_train = base_mean.predict(X_train)
y_hat_base_test = base_mean.predict(X_test)
```

```python
np.sqrt(metrics.mean_squared_error(y_train, y_hat_base_train))
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
# cross-validation of train test split

lr = linear_model.LinearRegression()
print(cross_val_score(lr, X_train, y_train, cv=3).mean())
print(cross_val_score(lr, X_test, y_test, cv=3).mean())
```

```python
# Instantiate Standard Scaler.
ss = StandardScaler()

# Fit AND transform Standard Scaler to overfit data in one step.
# Z_overfit_train = pd.DataFrame(ss_overfit.fit_transform(X_overfit_train))
Z_train = ss.fit_transform(X_train)
Z_test = ss.transform(X_test)
```

```python
# credits to lecture 4.05

# Instantiate our PolynomialFeatures object to create all two-way terms. 
poly = PolynomialFeatures(degree=2, interaction_only=False, include_bias=False)

# Fit and transform our X data.
X_overfit = poly.fit_transform(X)
```

```python
# Instantiate linear regression model without regularization.
lr_model = LinearRegression()
```

```python
# Create train/test splits.
Z_train, Z_test, y_train, y_test = train_test_split(
    X_overfit,
    y,
    test_size=0.7,
    random_state=42
)
```

```python
# Fit our model.
lr_model.fit(Z_train, y_train)
```

```python
Z_train.shape
```

```python
y_train.shape
```

```python
# Calculate R2 to see if our model is overfit, underfit, or neither.
print(lr_model.score(Z_train, y_train))
print(lr_model.score(Z_test, y_test))
```

The R2 value is lower on the test set so our model is Overfit.

# Ridge Regression

```python
# Set up a list of ridge alphas to check.
r_alphas = np.logspace(0, 5, 100)
# Generates 100 values equally between 0 and 5,
# then converts them to alphas between 10^0 and 10^5.

# Cross-validate over our list of ridge alphas.
ridge_model = RidgeCV(alphas=r_alphas, scoring='r2', cv=5)

# Fit model using best ridge alpha!
ridge_model = ridge_model.fit(Z_train, y_train)
```

```python
#Optimal value of alpha
ridge_model.alpha_
```

```python
#corresponding value of R2
print(ridge_model.score(Z_train, y_train))
print(ridge_model.score(Z_test, y_test))
```

The R2 value is lower on the test set so our model is Overfit.
